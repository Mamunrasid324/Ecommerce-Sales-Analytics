```python
sales_df = pd.DataFrame({
    "OrderID": range(5001, 7001),
    "CustomerID": np.random.choice(customers_df["CustomerID"], 2000),
    "ProductID": np.random.choice(products_df["ProductID"], 2000),
    "Quantity": np.random.randint(1, 6, 2000),
    "OrderDate": np.random.choice(
        pd.date_range("2023-01-01", "2025-12-31"),
        2000
    )
})

sales_df.head()
```

```python
sales_df = sales_df.merge(
    products_df[["ProductID", "Price"]],
    on="ProductID",
    how="left"
)

sales_df.head()
```

```python
sales_df.info()
```

```python
sales_df.shape
```

```python
sales_df.columns
```

```python
sales_df["SalesAmount"] = sales_df["Quantity"] * sales_df["Price"]
```

```python
sales_df.columns
```

```python
total_revenue = sales_df["SalesAmount"].sum()
print(total_revenue)
```

```python
total_orders = sales_df["OrderID"].count()
print("Total Orders:", total_orders)
```

```python
average_order_value = sales_df["SalesAmount"].mean()
print("Average Order Value:", average_order_value)
```

```python
sales_df["Month"] = sales_df["OrderDate"].dt.month_name()
```

```python
monthly_sales = sales_df.groupby("Month")["SalesAmount"].sum()

monthly_sales
```

```python
import matplotlib.pyplot as plt

monthly_revenue.plot(figsize=(12,5))

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.savefig("monthly_revenue.png", dpi=300, bbox_inches="tight")
plt.show()
```
